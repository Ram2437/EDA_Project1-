```python
import pandas as pd

import matplotlib.pyplot as plt

import numpy as np
from sklearn import linear_model

%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\ramakrishna.n\Data_sets\Canada_Per_capita_income.csv')
df.head()
```

```python
df['percapitaincome']= df['per capita income (US$)']
df.head()
```

```python
df.dtypes
```

```python
plt.xlabel("Year")
plt.ylabel("percapitaincome($)")
plt.scatter(df.year,df.percapitaincome,color='Red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.percapitaincome)
```

```python
t = reg.predict([[2020]])
t.dtype
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
df2 = pd.DataFrame({"year":[2017,2018,2019,2020], 
                    "percapitaincome":[reg.predict([[2017]]),reg.predict([[2018]]),reg.predict([[2019]]),reg.predict([[2020]])]})
# df2['percapitaincome'] = df2['percapitaincome'].astype(int)
df2
```

```python
df2.dtypes
```

```python
c = df2['percapitaincome'].astype(float)
e = df2['year']
```

```python
df3 = pd.concat([c, e], axis=1)
df3
```

```python
FinalDF = df.append(df3,sort=False,ignore_index=True)
FinalDF
```

```python
FinalDF['per capita income (US$)'] = FinalDF['percapitaincome'].values
FinalDF
```

```python
plt.xlabel("year")
plt.ylabel("percapitaincome")
plt.scatter(FinalDF.year,FinalDF.percapitaincome,color='Red', marker='+')
plt.plot(FinalDF.year,reg.predict(FinalDF[['year']]),color='blue')
```

```python
FinalDF.to_csv("prediction_CanadaPercapita.csv",index=False)
```
